```python
# Cell 1 - Mount ADLS Gen2 Storage
storage_account_name = "deprojectstorage1"
storage_account_key = "YOUR_STORAGE_ACCOUNT_KEY"
container_name = "raw"

spark.conf.set(
    f"fs.azure.account.key.{storage_account_name}.dfs.core.windows.net",
    storage_account_key
)

print("Storage mounted successfully!")
```

```python
# Cell 2 - Read CSV from ADLS
file_path = f"abfss://{container_name}@{storage_account_name}.dfs.core.windows.net/US_Accidents_March23.csv"

df = spark.read.csv(file_path, header=True, inferSchema=True)

print(f"Total rows: {df.count():,}")
print(f"Total columns: {len(df.columns)}")
```

```python
# Cell 3 - Basic Exploration
df.printSchema()
```

```python
# Cell 4 - Show sample data
display(df.limit(10))
```

```python
# Cell 5 - Null value analysis
from pyspark.sql.functions import col, sum as spark_sum

null_counts = df.select([
    spark_sum(col(c).isNull().cast("int")).alias(c) 
    for c in df.columns
]).toPandas().T

null_counts.columns = ["null_count"]
null_counts["null_percent"] = (null_counts["null_count"] / df.count() * 100).round(2)
null_counts = null_counts[null_counts["null_count"] > 0].sort_values("null_percent", ascending=False)

print(null_counts)
```

```python
# Cell 6 - Key statistics
print("Accidents by Severity:")
display(df.groupBy("Severity").count().orderBy("Severity"))

print("Top 10 States by Accidents:")
display(df.groupBy("State").count().orderBy("count", ascending=False).limit(10))

print("Date range of dataset:")
df.selectExpr("min(Start_Time)", "max(Start_Time)").show()
```
